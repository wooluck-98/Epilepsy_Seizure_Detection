# data
- The data set CHB-MIT contains 686 EEG recordings from 23 patients of different ages ranging from 1.5 years to 22 years. The recordings include 198 seizures. The used sampling frequency is 256 Hz.
- And some data files, for example, Chb12_27.edf, Chb12_28.edf and Chb12_29.edf, have different channel montages from other seizure files. In our experiments, we did not use the data in the above three EDF files. 

# data Segmentation 
- the data of 17 common channels were used for seizure/non-seizure features extraction
- channels : [P4-O2, FP2-F4, P7-O1, C4-P4, F7-T7, C3-P3, FP1-F7, F8-T8, FZ-CZ, CZ-PZ, F3-C3,
              T7-P7, P8-O2, FP1-F3, F4-C4, FP2-F8, P3-O1]
- length of 23 seconds from the beginning to the end without overlapping
- if a segment contained a seizure, it was considered as a seizure segment otherwise, it was a non-seizure segment
- In the seizure segments, the lengths of seizure data varied from 1 to 23s, with the average length being 16.9s
- As a result of the splitting, 665 seizure segments were obtained
- For evaluation over a balanced data, 665 non-seizure segments in each experiment were randomly selected from all the non-seizure segments without using random seed.

```python
channels_list = channels[:14] + channels[16:19]
channels_list.sort()
channels_list
```

```python
#seizure time
import re

seizure_file = []
seizure_time = []
with open("MIT_CHB/RECORDS-WITH-SEIZURES.txt", 'r') as txt_f:
    content = txt_f.readlines()
    for i in content[:-1]:
        ch = i.replace("\n", "").split("/")[0]
        
        #ch12는 제외
        if ch[-2:] == "12":
            continue
            
        ch_series = i.replace("\n", "").split("/")[1]
        seizure_file.append(ch_series)
        annotation_file = ch + '/' + ch + '-summary.txt'
        with open("MIT_CHB/" + annotation_file, 'r') as summary_f:
            content2 = summary_f.readlines()
            for j in content2:
                if "Seizure Start Time" in j:
                    start_time = int(re.findall("\d+", j)[0])
                if "Seizure End Time" in j:
                    end_time = int(re.findall("\d+", j)[0])
                    seizure_time.append((start_time, end_time))

#result
for i in range(len(seizure_file)):
    print(seizure_file[i],":", seizure_time[i])
```

```python
#data segment
import numpy as np
non_seizure_seg = [[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
seizure_seg = [[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
with open("MIT_CHB/RECORDS.txt", 'r') as records_f:
    content = records_f.readlines()
    for i in content[:6]:
        file = "MIT_CHB" + "/" + i.replace("\n", "")
        
        # chb12_27 / chb12_28 / chb12_29 제외
        if file.split("/")[2] == "chb12_27.edf" or file.split("/")[2] == "chb12_28.edf" or file.split("/")[2] == "chb12_29.edf":
            continue
            
        data = mne.io.read_raw_edf(file)
        raw_data = data.get_data()
        channels = data.ch_names
        for j in range(len(channels)):
            if channels[j] in channels_list:
                ch_idx = channels_list.index(channels[j])
                # seizure segment
                if file.split("/")[2] in seizure_file:
                    seizure_idx = seizure_file.index(file.split("/")[2])
                    start_time = seizure_time[seizure_idx][0]
                    end_time = seizure_time[seizure_idx][1]
                    for k in range(raw_data.shape[1]//256//23):
                        if start_time <= k*23 <= end_time:
                            seizure_seg[ch_idx].append(raw_data[ch_idx][k*23*256 : (k+1)*23*256])
                        else:
                            non_seizure_seg[ch_idx].append(raw_data[ch_idx][k*23*256 : (k+1)*23*256])
                else:
                    for k in range(raw_data.shape[1]//256//23):
                        non_seizure_seg[ch_idx].append(raw_data[ch_idx][k*23*256 : (k+1)*23*256])

print(len(seizure_seg), len(non_seizure_seg))
```
